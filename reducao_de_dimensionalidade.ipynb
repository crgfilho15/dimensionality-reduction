```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('Lenna.png')
```

```python
cv2_imshow(img)
```

```python
img_gray = img

'''for i in range(0, img.shape[0]):
  for j in range(0, img.shape[1]):
    img[i][j][0] = 0.299*img[i][j][0]
    img[i][j][1] = 0.587*img[i][j][1]
    img[i][j][2] = 0.114*img[i][j][2]'''

'''for i in range(0, img.shape[0]):
  for j in range(0, img.shape[1]):
    img[i][j] = (0.299*img[i][j][0], 0.587*img[i][j][1], 0.114*img[i][j][2])'''

# Apliquei a fórmula (L = 0.299*R + 0.587*G + 0.114*B) para se obter a luminância da imagem
# Fonte da equação: https://www.inf.ufrgs.br/~cwaraujo/inf01046/trabalho1.html

for i in range(0, img.shape[0]):
  for j in range(0, img.shape[1]):
    img_gray[i][j] = (0.299*img[i][j][0], 0.587*img[i][j][1], 0.114*img[i][j][2])

for i in range(0, img.shape[0]):
  for j in range(0, img.shape[1]):
    img_gray[i][j] = ((img[i][j][0] + img[i][j][1] + img[i][j][2]))

cv2_imshow(img_gray)
```

```python
img_bw = img_gray

for i in range(0, img_gray.shape[0]):
  for j in range(0, img_gray.shape[1]):
    if img_gray[i][j][0] <= 120:
      img_bw[i][j] = 0
    else:
      img_bw[i][j] = 255

cv2_imshow(img_bw)
```
